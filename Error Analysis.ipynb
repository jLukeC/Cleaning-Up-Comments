# Examples of each label


```python
import pandas as pd

labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

train = pd.read_csv('data/train.csv')
```

```python
def print_comments(comments):
    for index, comment in comments.iterrows():
        print(comment[labels])
        print(comment["comment_text"])
        print("------------------")
```

# Random example of each label

```python
for label in labels:
    print_comments(train[train[label] == 1].sample(1))
```

```python
test_BiLSTM = pd.read_csv('submissions/test_bidirectional_LSTM.csv')
test = pd.read_csv('data/test.csv')
```

# Misclassifications by the BiLSTM at 0.2 threshold

```python
test_BiLSTM_at_threshold = (test_BiLSTM[labels] + 0.2).astype(int)
```

```python
predicted_positives = test_BiLSTM_at_threshold.sum(axis=1)
actual_positives = test.sum(axis=1)
```

```python
false_positives = test[predicted_positives > actual_positives]
false_negatives = test[predicted_positives < actual_positives]
```

```python
false_positives["PREDICTED"] = predicted_positives
false_positives["ACTUAL"] = actual_positives

false_negatives["PREDICTED"] = predicted_positives
false_negatives["ACTUAL"] = actual_positives
```

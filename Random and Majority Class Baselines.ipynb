# Random Baseline
This script outputs a random prediction for every label in each instance

```python
import pandas as pd
test = pd.read_csv('data/test.csv')

labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
```

```python
import numpy as np

for label in labels:
    test[label] = np.random.random_sample(len(test[label]))
```

```python
test.sample(10)
```

```python
test.to_csv('submissions/test_random_baseline.csv', index=False)
```
